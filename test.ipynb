```python
import pandas as pd
import requests
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import os
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

# Create a directory to save the images
output_dir = 'album_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Step 2: Define a function to download and save the album image
def download_album_image(track_url, album_name):
        # Set up the WebDriver (ensure ChromeDriver is in your PATH or specify the path)
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service)

        # Open the Spotify web page (replace with the actual URL if needed)
        driver.get(track_url)

        # Wait for the page to load completely (adjust as needed)
        time.sleep(5)

        # Try to find the image element and get its URL
        try:
            img_element = driver.find_element(By.CSS_SELECTOR, ".mMx2LUixlnN_Fu45JpFB.CmkY1Ag0tJDfnFXbGgju._EShSNaBK1wUIaZQFJJQ.Yn2Ei5QZn19gria6LjZj")
            img_url = img_element.get_attribute("src")
            print(f"Image URL: {img_url}")
            
            # Download the image and save it locally
            response = requests.get(img_url)
            response.raise_for_status()  # Check for request errors

            # Create a folder to save the image if it doesn't exist
            output_dir = 'album_images'
            if not os.path.exists(output_dir):
                os.makedirs(output_dir)

            # Save the image with the album name (e.g., 'AlbumName.jpg')
            image_path = os.path.join(output_dir, f"{album_name}.jpg")

            with open(image_path, 'wb') as file:
                file.write(response.content)
            print(f"Image saved to {image_path}")
        except Exception as e:
            print(f"Error finding image: {e}")

        # Close the browser window
        driver.quit()
        return None

# Step 3: Iterate through the DataFrame and download the images
image_paths = []
for index, row in track.iterrows():
    album_name = row['Album']  # Assuming there's a column named 'Album'
    track_url = row['Uri']  # Assuming there's a column named 'Url_spotify'
    
    # Call the function to download and save the image
    image_path = download_album_image(track_url, album_name)
    
    if image_path:
        image_paths.append(image_path)
    else:
        image_paths.append(None)

# Step 4: Add image paths to the DataFrame
track['Album_Image_Path'] = image_paths

# Step 5: Save the DataFrame to a CSV
track.to_csv('album_img.csv', index=False)
```
