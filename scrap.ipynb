```python
import pandas as pd
import requests
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import os
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

# Create a directory to save the images
output_dir = 'album_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Step 2: Define a function to download and save the album image
def download_album_image(track_url, album_name):
        # Set up the WebDriver (ensure ChromeDriver is in your PATH or specify the path)
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service)

        # Open the Spotify web page (replace with the actual URL if needed)
        driver.get(track_url)

        # Wait for the page to load completely (adjust as needed)
        time.sleep(5)

        # Try to find the image element and get its URL
        try:
            img_element = driver.find_element(By.CSS_SELECTOR, ".mMx2LUixlnN_Fu45JpFB.CmkY1Ag0tJDfnFXbGgju._EShSNaBK1wUIaZQFJJQ.Yn2Ei5QZn19gria6LjZj")
            img_url = img_element.get_attribute("src")
            print(f"Image URL: {img_url}")
            
            # Download the image and save it locally
            response = requests.get(img_url)
            response.raise_for_status()  # Check for request errors

            # Create a folder to save the image if it doesn't exist
            output_dir = 'album_images'
            if not os.path.exists(output_dir):
                os.makedirs(output_dir)

            # Save the image with the album name (e.g., 'AlbumName.jpg')
            image_path = os.path.join(output_dir, f"{album_name}.jpg")

            with open(image_path, 'wb') as file:
                file.write(response.content)
            print(f"Image saved to {image_path}")
        except Exception as e:
            print(f"Error finding image: {e}")

        # Close the browser window
        driver.quit()
        return None

# Step 3: Iterate through the DataFrame and download the images
image_paths = []
for index, row in track.iterrows():
    album_name = row['Album']  # Assuming there's a column named 'Album'
    track_url = row['Uri']  # Assuming there's a column named 'Url_spotify'
    
    # Call the function to download and save the image
    image_path = download_album_image(track_url, album_name)
    
    if image_path:
        image_paths.append(image_path)
    else:
        image_paths.append(None)

# Step 4: Add image paths to the DataFrame
track['Album_Image_Path'] = image_paths

# Step 5: Save the DataFrame to a CSV
track.to_csv('album_img.csv', index=False)
```

```python
import os
import yt_dlp
from basic_pitch.inference import predict_and_save
from basic_pitch.models import model
import tensorflow as tf
import traceback

# tf.compat.v1.disable_eager_execution()  # For compatibility with tf.placeholder/


def download_audio_with_ytdlp(url, song_name, output_folder='sounds'):
    """
    Download and save audio using yt-dlp.
    Args:
        url (str): YouTube video URL.
        song_name (str): Name of the song (used for naming).
        output_folder (str): Directory to save the audio files.
    Returns:
        str: Path to the saved audio file, or None if failed.
    """
    try:
        os.makedirs(output_folder, exist_ok=True)
        output_template = os.path.join(output_folder, f"{song_name}.%(ext)s")
        
        ydl_opts = {
            'format': 'bestaudio/best',
            'outtmpl': output_template,
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'wav',
            }]
        }
        
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        
        print(f"Audio saved for: {song_name}")
        return os.path.join(output_folder, f"{song_name}.wav")
    
    except Exception as e:
        print(f"Error downloading {url}: {e}")
        return None


def convert_audio_to_midi(audio_path, song_name, output_folder='midi_files'):
    """
    Convert an audio file to MIDI using Basic Pitch.
    Args:
        audio_path (str): Path to the audio file.
        output_folder (str): Directory to save the MIDI files.
    Returns:
        str: Path to the saved MIDI file, or None if failed.
    """
    try:
        os.makedirs(output_folder, exist_ok=True)
        midi_path = os.path.join(output_folder, f"{song_name}.mid")
        
        my_model = model()
        
        
        # Call the predict_and_save function with correct arguments
        predict_and_save(
            audio_path_list= [audio_path],  # List of audio paths
            save_midi=True,            # Save the MIDI output
            sonify_midi=False,         # Do not create an audio preview of the MIDI
            save_model_outputs=False,   # Save model outputs
            save_notes=True,           # Save the detected notes as text
            model_or_model_path='./model',  # Use the default model
            output_directory=output_folder      # Path to save the MIDI file
        )
        print(f"MIDI saved for: {audio_path}")
        return midi_path
    
    except Exception as e:
        print(f"Error converting {audio_path} to MIDI: {e}")
        print(traceback.format_exc())
        return None


# Ensure you have loaded your DataFrame `song_track`
# Make sure 'song_track' DataFrame has 'Track' and 'Url_youtube' columns

# Download and convert each track
audio_paths = []
midi_paths = []

for _, row in song_track.iterrows():
    song_name = row['Track']
    track_url = row['Url_youtube']
    
    # Download audio
    audio_path = download_audio_with_ytdlp(track_url, song_name)
    if audio_path:
        audio_paths.append(audio_path)
        
        # Convert to MIDI
        midi_path = convert_audio_to_midi(audio_path, song_name=song_name)
        if midi_path:
            midi_paths.append(midi_path)

print("All downloads and conversions completed.")
print("Audio files:", audio_paths)
print("MIDI files:", midi_paths)
```
